```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import RandomOverSampler, SMOTE, ADASYN
from imblearn.under_sampling import RandomUnderSampler, NearMiss

data = pd.read_csv("Creditcard_data.csv")

X = data.drop("Class", axis=1)
y = data["Class"]

ros = RandomOverSampler(random_state=42)
X_bal, y_bal = ros.fit_resample(X, y)

models = []
models.append(LogisticRegression(max_iter=1000))
models.append(DecisionTreeClassifier())
models.append(KNeighborsClassifier())
models.append(RandomForestClassifier())
models.append(SVC())

sampling_methods = []
sampling_methods.append(RandomOverSampler(random_state=42))
sampling_methods.append(RandomUnderSampler(random_state=42))
sampling_methods.append(SMOTE(random_state=42))
sampling_methods.append(ADASYN(random_state=42))
sampling_methods.append(NearMiss())

results = pd.DataFrame(index=["M1","M2","M3","M4","M5"],
                       columns=["Sampling1","Sampling2","Sampling3","Sampling4","Sampling5"])

for i in range(5):
    sampler = sampling_methods[i]
    X_res, y_res = sampler.fit_resample(X_bal, y_bal)
    X_train, X_test, y_train, y_test = train_test_split(X_res, y_res, test_size=0.3, random_state=42)

    for j in range(5):
        model = models[j]
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        acc = accuracy_score(y_test, y_pred) * 100
        results.iloc[j, i] = round(acc, 2)

print(results)

for j in range(5):
    best_sampling = results.iloc[j].astype(float).idxmax()
    best_accuracy = results.iloc[j].astype(float).max()
    print(results.index[j], "best with", best_sampling, "Accuracy:", best_accuracy)
```
